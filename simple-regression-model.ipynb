```python
stocks = "GOOG SPY".split()
start = dt.date.today() - dt.timedelta(365)
```

```python
import yfinance as yf

data = yf.download(stocks, start=start)
data_close = data['Close']
print(data_close.head())
data.isnull().sum()
```

```python
returns = (np.log(data_close).diff()).dropna()
returns.head()
```

```python
returns.corr()
```

```python
sample = returns.sample(60).corr()
sample
```
